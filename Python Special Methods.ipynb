# 1. Constructor (`__init__()`)

## Definition

 `__init__()` method is a special method in Python that is automatically called whenever a new object is created. It is mainly used to initialize the attributes of an object.

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

book1 = Book("Electrical Machines", "Chapman")

print(book1.title)
print(book1.author)
```

# 2. String Representation (`__str__()`)

## Definition

`__str__()` method returns a user-friendly string representation of an object. It is automatically called when the `print()` function is used on an object.

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student Name: {self.name}"

s1 = Student("Nathan")

print(s1)
```

# 3. Developer Representation (`__repr__()`)

## Definition

`__repr__()` method returns an official string representation of an object. It is mainly intended for developers and debugging. If `__str__()` is not defined, Python uses `__repr__()` when the object is printed.

```python
class Student:
    def __init__(self, name, level):
        self.name = name
        self.level = level

    def __repr__(self):
        return f"Student(name='{self.name}', level={self.level})"

s1 = Student("Nathan", 100)

print(repr(s1))
```

# 4. Length (`__len__()`)

## Definition

 `__len__()` method allows an object to work with the `len()` function. It returns the number of items stored in the object.

```python
class Playlist:
    def __init__(self, songs):
        self.songs = songs

    def __len__(self):
        return len(self.songs)

playlist = Playlist(["Song 1", "Song 2", "Song 3"])

print(len(playlist))
```

# 5. Equality Comparison (`__eq__()`)

## Definition

 `__eq__()` method defines how two objects are compared using the `==` operator. It returns `True` if the objects are considered equal and `False` otherwise.

```python
class Student:
    def __init__(self, student_id):
        self.student_id = student_id

    def __eq__(self, other):
        return self.student_id == other.student_id

s1 = Student(101)
s2 = Student(101)

print(s1 == s2)
```

# 6. Addition (`__add__()`)

## Definition

 `__add__()` method overloads the addition (`+`) operator. It defines what should happen when two objects are added together.

```python
class Wallet:
    def __init__(self, amount):
        self.amount = amount

    def __add__(self, other):
        return Wallet(self.amount + other.amount)

    def __str__(self):
        return f"Balance: GHS {self.amount}"

w1 = Wallet(500)
w2 = Wallet(250)

print(w1 + w2)
```

# 7. Subtraction (`__sub__()`)

## Definition

`__sub__()` method overloads the subtraction (`-`) operator. It defines how one object is subtracted from another.

```python
class Vector:
    def __init__(self, x):
        self.x = x

    def __sub__(self, other):
        return Vector(self.x - other.x)

    def __str__(self):
        return str(self.x)

v1 = Vector(10)
v2 = Vector(3)

print(v1 - v2)
```

# 8. Multiplication (`__mul__()`)

## Definition

`__mul__()` method overloads the multiplication (`*`) operator. It defines what happens when two objects are multiplied.

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def __mul__(self, other):
        return (self.length * self.width) * (other.length * other.width)

r1 = Rectangle(4, 5)
r2 = Rectangle(2, 3)

print(r1 * r2)
```

# 9. Item Access (`__getitem__()`)

## Definition

`__getitem__()` method allows an object to be accessed using square brackets (`[]`). It makes a custom object behave like a list or dictionary.

```python
class Library:
    def __init__(self, books):
        self.books = books

    def __getitem__(self, index):
        return self.books[index]

library = Library(["Physics", "Mathematics", "Programming"])

print(library[1])
```

# 10. Item Assignment (`__setitem__()`)

## Definition

 `__setitem__()` method allows values in an object to be assigned or modified using square brackets (`[]`).

```python
class Scores:
    def __init__(self):
        self.marks = {}

    def __setitem__(self, student, score):
        self.marks[student] = score

scores = Scores()

scores["Nathan"] = 89
scores["Ama"] = 95

print(scores.marks)
```

# 11. Membership Testing (`__contains__()`)

## Definition

 `__contains__()` method is used with the `in` keyword. It checks whether a value exists in an object and returns `True` or `False`.

```python
class Classroom:
    def __init__(self, students):
        self.students = students

    def __contains__(self, name):
        return name in self.students

room = Classroom(["Eric", "Kofi", "Ama"])

print("Ama" in room)
print("John" in room)
```

# 12. Callable Objects (`__call__()`)

## Definition

 `__call__()` method allows an object to behave like a function. After defining this method, an object can be called using parentheses `()`.

```python
class Multiplier:
    def __init__(self, number):
        self.number = number

    def __call__(self, value):
        return self.number * value

double = Multiplier(2)

print(double(15))
```

# 13. Iteration (`__iter__()`)

## Definition

 `__iter__()` method returns an iterator object. It allows an object to be used in a `for` loop.

```python
class Numbers:
    def __init__(self):
        self.values = [10, 20, 30, 40]

    def __iter__(self):
        return iter(self.values)

nums = Numbers()

for number in nums:
    print(number)
```

# 14. Next Item (`__next__()`)

## Definition

 `__next__()` method returns the next item in an iterator. When there are no more items, it raises `StopIteration`.

```python
class Countdown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.current == 0:
            raise StopIteration

        number = self.current
        self.current -= 1
        return number

for i in Countdown(5):
    print(i)
```

# 15. Boolean Evaluation (`__bool__()`)

## Definition

 `__bool__()` method determines whether an object is considered `True` or `False` when used in a Boolean expression.

```python
class Battery:
    def __init__(self, percentage):
        self.percentage = percentage

    def __bool__(self):
        return self.percentage > 0

phone1 = Battery(80)
phone2 = Battery(0)

print(bool(phone1))
print(bool(phone2))
```

# 16. Destructor (`__del__()`)

## Definition

 `__del__()` method is called automatically when an object is about to be destroyed. It is mainly used to perform cleanup operations before the object is removed from memory.

```python
class Student:
    def __init__(self, name):
        self.name = name
        print(f"{self.name} has been created.")

    def __del__(self):
        print(f"{self.name} has been deleted.")

s1 = Student("Nathan")

del s1
```
